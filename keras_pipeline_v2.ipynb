```python
import os
import keras
from keras.preprocessing.text import Tokenizer
import tensorflow as tf

lang_exts_with_dot = ['.html', '.java', '.py', '.c', '.cpp', '.rb', '.php']

lang_count = {
    '.html' : 0, '.java': 0, '.py': 0, '.c': 0, '.cpp': 0, '.rb': 0, '.php': 0
}

X_raw = []
labels_raw = []
counter = 0
for root, subdir, files in os.walk('./code-repos/'):
    for file in files:
        try:
            ext = os.path.splitext(file)[-1].lower()
            if not ext in lang_exts_with_dot:
                continue
            filepath = os.path.join(root, file)
            f = open(filepath, 'rb')
            text = f.read().decode(errors='replace')
            text = text[:1028]
            print(counter)
            counter += 1
            X_raw.append(text)
            labels_raw.append(ext)
            lang_count[ext] += 1
            if lang_count[ext] == 2000:
                lang_exts_with_dot.remove(ext)
            if counter == 14000:
                break 
        except:
            continue
    if counter == 14000:
        break
        
print("A")
tokenizer = Tokenizer(num_words=9999, char_level = True)
print("B")
tokenizer.fit_on_texts(X_raw)
print("C")
sequences = tokenizer.texts_to_sequences(X_raw)
print("D")

#look at pre padding.
seq = keras.preprocessing.sequence.pad_sequences(sequences, maxlen=None, dtype='int32', padding='pre', value=0.0)
print("E")


from sklearn.preprocessing import OneHotEncoder 

print(seq.flatten())

# seq = seq.reshape(-1,1)

# print(tokenizer.get_config())
print("F")
enc = OneHotEncoder()
depth = 56
print("G")

X_encoded = tf.one_hot(seq, depth,
                       
           axis=-1)

# X_encoded = enc.fit_transform(seq.flatten())

print(X_encoded.shape)


#One hot encode each of the list independently from one another. Output shape should be 3 dimensional. 

# print(tokenizer.get_config())

print(type(X_encoded))

print(set(labels_raw))
```

```python
from keras.utils import to_categorical
from numpy import array
import pandas as pd
from sklearn import preprocessing

# data = array(labels_raw)
# print(data)
# # one hot encode
# encoded = to_categorical(data)
# print(encoded)


le = preprocessing.LabelEncoder()

y = le.fit_transform(labels_raw)

print(y)

#convert .java, .c, .html ...
```

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.3, random_state=42)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
print(y_train)
print(y_test)
```

```python
from keras.models import Model
from keras.layers import Dense, Dropout, LSTM, Input, Activation
from keras import optimizers
from keras.callbacks import History 
from sklearn.metrics import mean_squared_error

lstm_input = Input(shape=(X_train.shape[1],56), name='input_for_lstm')

inputLSTM = LSTM(21, name='first_layer')(lstm_input)
inputsDense1 = Dense(32, name='first_dense_layer')(inputLSTM)
inputsDense2 = Dense(7, name='dense_layer')(inputsDense1)
output = Activation('softmax', name='output')(inputsDense2)

model3 = Model(inputs=lstm_input, outputs=output)
adam = optimizers.Adam(lr = 0.004)

model3.compile(optimizer=adam, loss='sparse_categorical_crossentropy')
# model.summary()

checkpoint_cb = keras.callbacks.ModelCheckpoint("my_keras_model.h5", save_best_only=True)

early_stopping_cb = keras.callbacks.EarlyStopping(
    monitor="val_loss",
    patience=10,
    mode="min",
    restore_best_weights=True)
history = model3.fit(x=X_train, y=y_train, batch_size=100, epochs=100, shuffle=True, validation_split = 0.1, callbacks=[checkpoint_cb, early_stopping_cb])
# model3.fit(x=X_train, y=y_train, batch_size=100, epochs=100, shuffle=True, validation_split = 0.1)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model3.predict(X_test)
print(y_pred)
y_pred = np.argmax(y_pred, axis=1)
print(accuracy_score(y_test, y_pred))
```
