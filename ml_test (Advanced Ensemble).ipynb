```python
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_csv('./processed_data/df_train.csv')
df_test = pd.read_csv('./processed_data/df_test.csv')
```

```python
df_train.head()
print(df_test.info)
```

```python
df_validation = pd.read_csv('./processed_data/df_valid.csv')

df_combined = pd.concat((df_validation, df_train))
X_train_validation = df_combined.drop('label', axis=1)
y_train_validation = df_combined['label']
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold
import statistics
from sklearn.metrics import accuracy_score

kf = KFold(n_splits = 3)

k_fold_accuracy = []

model = CatBoostClassifier(random_seed = 5, verbose = True, boosting_type = "Plain", learning_rate = 0.08, depth = 8)

for train_index, validation_index in kf.split(X_train_validation):
    X_train, X_valid = X_train_validation.iloc[train_index], X_train_validation.iloc[validation_index]
    y_train, y_valid = y_train_validation.iloc[train_index], y_train_validation.iloc[validation_index]
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_valid)
    
    accuracy = accuracy_score(y_pred, y_valid)
    
    k_fold_accuracy.append(accuracy)
    
    
print('The accuracy for each fold is:', k_fold_accuracy)
print('The average accuracy is:', statistics.mean(k_fold_accuracy))
    
    
    
```

```python
df_test = pd.read_csv('./processed_data/df_test.csv')
X_test = df_test.drop('label', axis=1)
y_test = df_test['label']

y_pred_test = model.predict(X_test)

print(accuracy_score(y_pred_test, y_test))
```
